# 1. Create a NumPy array of shape (5, 5) with random integers between 1 and 50.

```python
import numpy as np
arr = np.random.randint(1, 51, (5, 5))
```

```python
arr
```

# a. Find the row with the maximum sum.

```python
row_sums = arr.sum(axis=1)
row_sums
```

```python
max_row_index = np.argmax(row_sums)
max_row_index
```

```python
print('The row with the maximum sum is ',max_row_index)
```

# b. Replace all elements greater than the mean with 0

```python
mean_value = arr.mean()
mean_value
```

```python
arr[arr > mean_value] = 0
print("Array after replacement:\n", arr)
```

# 3. Load the dataset Iris data

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('iris')

df
```

# a. Display the first 10 rows and basic statistics.

```python
df.head(10)
```

```python
df.describe()
```

# b. Filter all rows where sepal_length > median and species is "versicolor".

```python
median_sepal_length = df['sepal_length'].median()
median_sepal_length
```

```python
filtered_df = df[(df['sepal_length'] > median_sepal_length) & (df['species'] == 'versicolor')]
filtered_df
```

# c. Group the data by species and compute the mean petal_length.

```python
mean_petal_length = df.groupby('species')['petal_length'].mean()
mean_petal_length
```

# 4. Add a new column named petal_ratio = petal_length / petal_width.

```python
df['petal_ratio'] = df['petal_length'] / df['petal_width']
df['petal_ratio']
```

```python
df
```

# a. Find the top 5 rows with the highest petal_ratio.

```python
top_5_ratio = df.sort_values(by='petal_ratio', ascending=False).head(5)
top_5_ratio
```

# 5. Using Matplotlib, create:
# a. A histogram of sepal_length with 15 bins.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.hist(df['sepal_length'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

# b. A scatter plot of sepal_length vs petal_length, color-coded by species

```python
plt.subplot(1, 2, 2)
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
for species, color in colors.items():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal_length'], subset['petal_length'], 
                label=species, color=color)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()

plt.tight_layout()
plt.show()
```
